```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
x, y = np.meshgrid(np.linspace(-1,1,10), np.linspace(-1,1,10))
d = np.sqrt(x*x+y*y)
sigma, mu = 1.0, 0.0
g = np.exp(-( (d-mu)**2 / ( 2.0 * sigma**2 ) ) )
print("2D Gaussian-like array:")
print(g)
```

```python

def gaussian(mu,sigma):
  x, y = np.meshgrid(np.linspace(-1,1,10), np.linspace(-1,1,10))
  d = np.sqrt(x*x+y*y)
  g = np.exp(-( (d-mu)*(d-mu) / ( 2.0 * sigma*sigma ) ) )
  return g
```

```python
def plot(g):
  N = 10
  X = np.linspace(-2, 2, N)
  Y = np.linspace(-2, 2, N)
  X, Y = np.meshgrid(X, Y)
  fig = plt.figure()
  ax2 = fig.add_subplot(2,1,2,projection='3d')
  t = ax2.contourf(X, Y, g, zdir='z', offset=0, cmap=cm.viridis)
  ax2.view_init(90, 270)

  ax2.grid(False)
  ax2.set_xticks([])
  ax2.set_yticks([])
  ax2.set_zticks([])
  ax2.set_xlabel(r'$x_1$')
  ax2.set_ylabel(r'$x_2$')
  return t
```

```python
def contourtoimage(g):
  plt.imshow(g, cmap='bone_r')
  plt.show()
```
